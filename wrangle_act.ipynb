# Tidness Issue

**Issue**

Doggo, puppo, pupper, fluffer all mean the same

- Doggo is another name for dog

- Pupper are small dogs, a domestic dog with loveable attributes

- Puppo are dogs with loveable attributes

- Fluffer is an extreme fury dog.

It mostly likely look like different stages of a dog. So we create a new variable 'dog_stage' to cateogorize it. 

**Define:**

Using string concatenation add the doggo, floofer, pupper, puppo columns together as one column then drop each of the 4 columns

```python
df1_clean[['text','doggo','floofer','pupper','puppo']].sample(20)
```

```python
# add to a single column named dog_stage
df1_clean['dog_stage'] = (df1_clean['doggo'].replace('None',"")+df1_clean['floofer'].replace('None',"")+
                   df1_clean['pupper'].replace("None","")+df1_clean['puppo'].replace('None',""))

# replace empty stage observation with nan
df1_clean['dog_stage'].replace(" ",np.nan,inplace=True)
```

```python
df1_clean[['dog_stage']].sample(10)
```

Some rows mixed together, rename rows with 2 dog stages concatenated together to mixed then drop the doggo,pupper,floofer and puppo columns

```python
df1_clean.loc[df1_clean.dog_stage == 'doggopupper', 'dog_stage'] = 'mixed'
df1_clean.loc[df1_clean.dog_stage == 'doggopuppo', 'dog_stage'] = 'mixed'
df1_clean.loc[df1_clean.dog_stage == 'doggofloofer', 'dog_stage'] = 'mixed'
```

```python
df1_clean['dog_stage'].replace(" ",np.nan,inplace=True)
```

```python
df1_clean.drop(columns=['doggo','floofer','pupper','puppo'],inplace = True)
```

**Test**
